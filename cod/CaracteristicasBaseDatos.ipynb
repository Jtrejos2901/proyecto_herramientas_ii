# Características generales de la base de datos 

Este código presenta estadísticos básicos de la base de datos

```python
# Importación de bibliotecas necesarias
import numpy as np  # Para cálculos numéricos
import pandas as pd  # Para manipulación y análisis de datos

# Ruta de acceso
file_path = 'C:/Users/monts/OneDrive/Documentos/UCR/herramientas ciencias de datos II/proyecto_herramientas_ii/data/diabetes.csv'

# Leer el archivo CSV en un DataFrame 
data = pd.read_csv(file_path)

data
```

```python
def caracteristicas_de_una_columna(dataframe, columna):
        caracteristicas_columna = dataframe[columna]
        caracteristicas = {}

        if caracteristicas_columna.dtype in ['int64', 'float64']:  
            caracteristicas['Media'] = round(caracteristicas_columna.dropna().mean(), 2)
            caracteristicas['Mediana'] = round(caracteristicas_columna.dropna().median(), 2)
            caracteristicas['Mínimo'] = round(caracteristicas_columna.dropna().min(), 2)
            caracteristicas['Máximo'] = round(caracteristicas_columna.dropna().max(), 2)
            caracteristicas['Varianza'] = round(caracteristicas_columna.dropna().var(), 2)
            
        
        conteo_valores = caracteristicas_columna.dropna().value_counts()
        max_valor = conteo_valores.max()
        nombres_max = conteo_valores[conteo_valores == max_valor].index.tolist()
        imprimir = ', '.join(str(nombre) for nombre in nombres_max) 
        caracteristicas['Moda'] = imprimir

        caracteristicas['Datos faltantes'] = caracteristicas_columna.isnull().sum()

        return caracteristicas
    
def caracteristicas_dataframe_columnas(dataframe):
    resultados = {}
    for columna in dataframe.columns:
        resultados[columna] = caracteristicas_de_una_columna(dataframe, columna)
    df_resultados = pd.DataFrame(resultados)
    return df_resultados
```

```python
df_resultados = caracteristicas_dataframe_columnas(data)
print(df_resultados)
```
